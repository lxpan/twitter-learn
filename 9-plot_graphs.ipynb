```python
import os
import json

data_folder = os.path.join(os.path.expanduser("~"), "Data", "research")
# results_filename= os.path.join(data_folder, "attribution_results.json")
results_filename= os.path.join(data_folder, "attribution_results_control.json")
```

```python
""" 
1. read json file as dict
2. import graphing libs
3. specify axes
4. plot graph
"""

%matplotlib inline

# import the plot modules
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# read prediction accuracies from JSON
with open(results_filename, 'r') as inf:
    scores = json.load(inf)

df = pd.DataFrame.from_dict(scores)
```

```python
df
```

```python
sns.set(style="whitegrid", color_codes=True)

violin = sns.violinplot(data=df)
ax = violin.axes
# add x/y labels
ax.set_xlabel("Number of authors")
ax.set_ylabel("Accuracy")
# set the range
ax.set_ylim([0,1])

# save plot to disk
# violin.savefig("python_violin_axes_corrected.png")
```
